```python
from typing import List, Tuple, Dict, Union

def stats(numbers: List[int]) -> Tuple[float, int, int]:
    average = sum(numbers) / len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)
    return (average, minimum, maximum)
```
